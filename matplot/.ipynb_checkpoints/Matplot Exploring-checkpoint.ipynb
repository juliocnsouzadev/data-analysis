# Matplot Exploring

```python
import matplotlib.pyplot as plt
```

### creating some data to play with

```python
r = range(1900, 2100,10)
years = [ item * 1 for item in r]



def calc(item):
    i = 1900;
    divValue = (item / i)
    
    if item > 2070:
        divValue = ((item * 0.99 )/i)
    
    oneBillion = 1000000000 
    timesPower = 1052631 ** divValue
    itemTimes = item * timesPower
    result = itemTimes / oneBillion;
    return result
   
                
population = [ calc(item)  for item in r]
```

### ploting

```python
plt.plot(years, population)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population Projections')
plt.yticks([0,2,4,6,8,10], ['0','2bi','4bi','6bi','8bi','10bi'])
plt.show()
```

```python
plt.scatter(years, population)
plt.xscale('log')
plt.show()
```

### Historgram

```python
plt.hist(population, 5)
plt.show()
```
